```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
import kagglehub
import os
import zipfile

path = kagglehub.dataset_download("teamincribo/cyber-security-attacks")
print("Path to dataset files:", path)
```

```python
# Find the CSV file within the downloaded directory

for root, dirs, files in os.walk(path):
    for file in files:
        if file.endswith(".csv"):
            csv_file_path = os.path.join(root, file)
            print(f"Found CSV file: {csv_file_path}")
            try:
              df = pd.read_csv(csv_file_path)
              print(df.head())
              break # Stop after finding the first CSV file
            except pd.errors.ParserError:
              print(f"Error: Could not parse {csv_file_path} as a CSV file.")
            except Exception as e:
              print(f"An unexpected error occurred: {e}")
    else:
        continue  # Continue to the next directory if no CSV is found
    break # Exit the outer loop if a CSV is found
else:
    print("Error: No CSV file found in the downloaded dataset.")
```

```python
# Use Simple Imputation to replace null values with the column mean values

from sklearn.impute import SimpleImputer

imputeMean = SimpleImputer(strategy='most_frequent')
df[['Malware Indicators', 'Alerts/Warnings', 'Proxy Information', 'Firewall Logs', 'IDS/IPS Alerts']] = imputeMean.fit_transform(df[['Malware Indicators', 'Alerts/Warnings', 'Proxy Information', 'Firewall Logs', 'IDS/IPS Alerts']])

df
```
